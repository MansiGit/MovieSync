```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
# import pandas as pd

# Read the log file
df = pd.read_csv('./logs_cdn/log1.txt', sep='|', engine='python', header=None,
                 names=['Timestamp', 'RoomCode', 'UserName', 'Action'])
df
```

```python
# Convert Timestamp to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'], utc=True)
df['Timestamp'] = df['Timestamp'].astype(int) // 10**6
df
```

```python
# Group by RoomCode and calculate average latency
import datetime
grouped = df.groupby('RoomCode')
count = 0

for name, group in grouped:
    
    if count==2: break
        
    count += 1
    group = group.sort_values('Timestamp')
    if str(group["RoomCode"].iloc[0]).strip()=="": continue
    
    action_df = group.groupby('Action')
    
    for action, g in action_df:
        set_users = set()
        latency_values = []
        prev = None
        for index, row in g.iterrows():
            if row['UserName'] in set_users:
                num_clients = len(set_users)
                set_users.clear()
                latency = (sum(latency_values[1:]) - latency_values[0]*len(latency_values[1:])) / (len(latency_values)-1)
                print(f'Act:{action} | RC: {prev["RoomCode"]} | Lat:{latency}ms | NumClients {num_clients}')
                latency = []
                prev = None
            
            set_users.add(row['UserName'])
            latency_values.append(row['Timestamp'])
            prev = row
        
                
            
```

```python
df = pd.read_csv('bufferLog.txt', sep='|', engine='python', header=None,
                 names=['Timestamp', 'RoomCode', 'UserName', 'BufferRate'])
df['Timestamp'] = pd.to_datetime(df['Timestamp'], utc=True)
df['Timestamp'] = df['Timestamp'].astype(int) // 10**6
```

```python
grouped = df.groupby('RoomCode')

for name, group in grouped:
    group = group.sort_values('Timestamp')
    username_df = group.groupby('UserName')
    
    for n, g in username_df:
        print(n)
        fig, ax = plt.subplots()
        ax.plot(g['Timestamp'], g['BufferRate'])
        ax.set_xlabel("Time")
        ax.set_ylabel("Buffer Rate")
```
